# Define a year as a "Superman year" whose films feature more Superman characters than Batman. How many years in film history have been Superman years?

```python
# Assuming you have a DataFrame named cast with columns "year" and "character"

# Filter for roles named "Superman" or "Batman"
superman_batman = cast[cast['character'].isin(['Superman', 'Batman'])]

# Group by year and character and count occurrences
character_counts = superman_batman.groupby(['year', 'character']).size().unstack().fillna(0)

# Find the years where Superman occurrences are more than Batman occurrences
superman_years = character_counts[character_counts['Superman'] > character_counts['Batman']]

# Count the number of Superman years
number_of_superman_years = len(superman_years)

print(f"There have been {number_of_superman_years} Superman years in film history.")
```

# Build a plot with a line for each rank n=1 through n=3, where the line shows what fraction of that rank's roles were 'actor' roles for each year in the history of film.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named cast with columns "year", "type", and "n"

# Filter for roles with n values from 1 to 3
filtered_roles = cast[cast['n'].isin([1, 2, 3])]

# Group by year, type, and n, then count roles
role_counts = filtered_roles.groupby(['year', 'n', 'type']).size().unstack('type').fillna(0)

# Calculate the fraction of actor roles for each n value
fractions = role_counts['actor'] / (role_counts['actor'] + role_counts['actress'])

# Plotting the fractions for each n value
fig, ax = plt.subplots(figsize=(12, 7))
for rank in [1, 2, 3]:
    fractions.xs(rank, level='n').plot(ax=ax, label=f'n={rank}')

plt.title("Fraction of 'Actor' Roles for Ranks n=1 through n=3 Over the History of Film")
plt.xlabel("Year")
plt.ylabel("Fraction of Actor Roles")
plt.legend(title="Rank (n value)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA

```python
import pandas as pd

# Read the CSV
release_dates = pd.read_csv('release_dates.csv')

# Print out the first few rows to inspect
print(release_dates.head())
```

```python
# import matplotlib.pyplot as plt

# Filter movies with "Christmas" in their title and released in the USA
christmas_movies = release_dates[
    (release_dates['title'].str.contains('Christmas', case=False)) & 
    (release_dates['country'] == 'USA')
]

# Convert 'date' column to datetime format
christmas_movies['date'] = pd.to_datetime(christmas_movies['date'], errors='coerce')

# Extract month
christmas_movies['month'] = christmas_movies['date'].dt.month

# Count movies released in each month
month_counts = christmas_movies['month'].value_counts().sort_index()

# Plot
month_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Months Movies with "Christmas" in Title are Released in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()
```

# Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA

# In which months do films with Judi Dench tend to be released in the USA?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have both cast and release_dates DataFrames loaded

# Filter the cast DataFrame for movies that Judi Dench acted in
judi_dench_movies = cast[cast['name'] == 'Judi Dench']

# Join with the release_dates DataFrame to get USA release dates
release_dates_judi = judi_dench_movies.merge(release_dates[release_dates['country'] == 'USA'], on=['title', 'year'])

# Convert the 'date' column to a datetime format
release_dates_judi['date'] = pd.to_datetime(release_dates_judi['date'])

# Extract month from the date
release_dates_judi['month'] = release_dates_judi['date'].dt.month

# Count the number of movies released in each month
month_counts = release_dates_judi['month'].value_counts().sort_index()

# Plot
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_counts.index = month_names[:len(month_counts)]
month_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Months in which movies with Judi Dench are released in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```
